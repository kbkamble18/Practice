```python
def print_space(space):
    if space > 0:
        print(" ", end="")
        print_space(space - 1)
def print_star(star):
    if star > 0:
        print("*", end="")
        print_star(star - 1)
def print_pyramid(n, current_row=1):
    if current_row > n:
        return
    spaces = n - current_row
    stars = 2 * current_row - 1
    print_space(spaces)
    print_star(stars)
    print()
    print_pyramid(n, current_row + 1)
n = 5
print_pyramid(n)
```

```python
 # Encapsulation - Restrict access to certain attributes(variables) or methods(function) to protect data and enforce control access

class Student:    # Class
    def __init__(self, full_name, class_grade, percentage):    # Method
        self.name = full_name    # attribute
        self.grade = class_grade   # attribute
        self.__percentage= percentage   # Encapsulate attribute "__" - Limits access
        print(f"{name} is in grade {grade}")

    def get_percentage(self):           #we need to define method in order to access encapsulated attribute
        return self.__percentage
        
    def student_details(self): # Encapsulate Method
        print(f"{self.name} is in grade {self.grade}")

# object = instance of class
name = input("Enter name for the student")
grade = input("Enter grade of the student")
percentage = int(input("Enter percentage"))

# Object
student1 = Student(name,grade,percentage)     # Object
student2 = Student('Vishakha',12,99)
# print(student1.name, student1.grade)

student1.student_details()
#print(student1.percentage)    #Error
#print(student1.__percentage)    #Error

print(student1.get_percentage())
```
